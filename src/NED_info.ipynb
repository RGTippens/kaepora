```python
from bs4 import BeautifulSoup
import urllib2
from astroquery.ned import Ned
import composite
```

```python
SN_Array = composite.grab("SELECT * from Supernovae inner join Photometry ON Supernovae.SN = Photometry.SN", multi_epoch = True, make_corr = False)
```

```python
event_dict = {}
for i in range(len(SN_Array)):
    if event_dict.get(SN_Array[i].name) is None:
        event_dict[SN_Array[i].name] = [SN_Array[i]]
    else:
        event_dict[SN_Array[i].name] = event_dict.get(SN_Array[i].name) + [SN_Array[i]]
```

```python
host_dict = {}
for SN_name in event_dict.keys():
    name = 'SN' + SN_name
    url = 'https://ned.ipac.caltech.edu/cgi-bin/objsearch?objname=%s&extend=no&hconst=73&omegam=0.27&omegav=0.73&corr_z=1&out_csys=Equatorial&out_equinox=J2000.0&obj_sort=RA+or+Longitude&of=pre_text&zv_breaker=30000.0&list_limit=5&img_stamp=YES' % name
    response = urllib2.urlopen(url)
    ned_html = response.read()
    soup = BeautifulSoup(ned_html, 'html.parser')
    ned_text = soup.get_text().split('\n')
    for i in range(len(ned_text)):
        if 'CROSS-IDENTIFICATIONS' in ned_text[i]:
            break
    cross_id_text = ned_text[i:i+5]
    for line in cross_id_text:
        if ':' in line:
            cut1 = line.split('SN')
            for el in cut1:
                if ':' in el:
                    galaxy = el.split(':')[0].strip()
                    host_dict[galaxy] = SN_name
                    print galaxy
```

```python
def simplify_ned_morphology(morph):
    if 'S' in morph or 'E' in morph:
        morph = morph.replace('?', '')
        morph = morph.replace('edge-on', '')
        morph = morph.replace('high-surface-brightness galaxy', '')
        morph = morph.replace('system with jet', '')
        morph = morph.replace('+', '')
        morph = morph.replace('-', '')
        morph = morph.replace('compact', '')
        morph = morph.replace('/', '')
        morph = morph.replace('with extended material', '')
        morph = morph.replace('^', '')
        morph = morph.replace("(R')", '')
        morph = morph.replace("(R)", '')
        morph = morph.replace("(rs)", '')
        morph = morph.replace("(r)", '')
        morph = morph.replace("(s)", '')
        morph = morph.replace("Im", '')
        morph = morph.replace("pec", '')
        morph = morph.replace("pair", '')
        morph = morph.replace("diffuse spiral", '')
    else:
        morph = ''
    #removes barred classification
    if 'S' in morph:
        morph = morph.replace("S00", 'S')
        morph = morph.replace("A00", '0')
        morph = morph.replace("B00", '0')
        morph = morph.replace("A", '')
        morph = morph.replace("B", '')
        morph = morph.replace("S   Irr", 'S')
    if morph.strip() == 'S':
        morph = 'Sp'
    if len(morph) > 0:
        return morph
    else:
        return None
    
```

```python
simp_SN_info_dict = {}
for sn in SN_info_dict.keys():
    morph = SN_info_dict[sn][2]
    if morph != None:
        new_morph = simplify_ned_morphology(morph)
    else:
        new_morph = None
    if new_morph != None:
        simp_SN_info_dict[sn] = new_morph.strip()
    else:
        simp_SN_info_dict[sn] = new_morph
```

```python
hub_queries_m10 = ["SELECT * from Spectra inner join Events ON Spectra.SN = Events.SN where phase < -7 and Hubble_Res < 0 and ((dm15_source >= .85 and dm15_source < 1.8) or (dm15_from_fits >= .85 and dm15_from_fits < 1.8))",
                      "SELECT * from Spectra inner join Events ON Spectra.SN = Events.SN where phase < -7 and Hubble_Res > 0 and ((dm15_source >= .7 and dm15_source < 1.8) or (dm15_from_fits >= .7 and dm15_from_fits < 1.8))"]
```

```python
composites_m10, sn_arrays_m10, boot_sn_arrays_m10 = kpora.make_composite(hub_queries_m10, boot=True, medmean=1, verbose=False, gini_balance=True, combine=True)
```

```python
kpora.set_min_num_spec(composites_m10, 5)
kplot.comparison_plot(composites_m10, scale_type=False, min_num_show=5, savename='../../Hubble_Residual_Paper/hubble_res_m10')
```

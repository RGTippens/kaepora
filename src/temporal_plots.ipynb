```python
def stacked_plot(composites):
    font1 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 20,
        }
    
    font2 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 10,
        }
    
    fig, ax = plt.subplots(1,1)
    ax.get_yaxis().set_ticks([])
#     plt.ylim([-311,12])
    plt.ylabel('Relative Flux + const.', fontdict = font1, fontsize = 20)
    plt.xlabel('Wavelength ' + "($\mathrm{\AA}$)", fontdict = font1, fontsize = 20)
    plt.xticks(fontsize = 10)
    ax.xaxis.set_tick_params(length=5, width = 2)
    fig.set_size_inches(10.5, 13.5, forward = True)
#     fig.set_size_inches(20., 30., forward = True)
    composites, scales = composite.optimize_scales(composites, composites[0], True)

    i = 0
    for comp in composites:
        phase = np.average(comp.phase_array[comp.x1:comp.x2])
#         buff = 200*np.log10(phase+20)
        buff  = 5*i
        ax.plot(comp.wavelength[comp.x1:comp.x2], 2.e15*comp.flux[comp.x1:comp.x2] - buff, color = '#3F5D7D', linewidth = 2)
#         plt.fill_between(comp.wavelength[comp.x1:comp.x2], 2.e15*comp.low_conf[comp.x1:comp.x2]- 3.5*phase, 2.e15*comp.up_conf[comp.x1:comp.x2] - 3.5*phase, alpha = 0.5)
        plt.text(10300, 2.*comp.flux[comp.x2] - buff+ 1, " %+2.2f" % (phase), fontdict = font2)
#         plt.title('All Phase Composite Spectra', fontdict = font1, fontsize = 40)
        i+=1
#     plt.savefig('../../Paper_Drafts/all_t_spec_eq_spaced.png', dpi = 300, bbox_inches = 'tight')
    plt.show()
```

## Zoom In Plots

```python
import matplotlib.gridspec as gridspec
```

```python
def zoom_in_plot(lst):
    font1 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 20,
        }
    
    font2 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 10,
        }
    
    font3 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 20,
        }
    
    ca_si_range = [3400., 4400.]
    si_range = [5500., 6500.]
    nir_trip_range = [7500., 9000]
    
    gs = gridspec.GridSpec(1, 3)
    fig = plt.figure(num = 1, dpi = 100, figsize = [8.5,11], facecolor = 'w')
    fig.text(0.5, 0.04, 'Wavelength ' + "($\mathrm{\AA}$)", ha='center', fontdict = font3)
#     s_m = query_db.make_colorbar(lst)
    
    params = []
    for comp in lst:
        params.append(np.average(comp.phase_array[comp.x1:comp.x2]))

    norm = matplotlib.colors.Normalize(vmin=np.min(params),vmax=np.max(params))
    # c_m = matplotlib.cm.viridis
    c_m = matplotlib.cm.winter_r
    s_m = matplotlib.cm.ScalarMappable(cmap=c_m, norm=norm)
    s_m.set_array([])
    
    i = 0
    for comp in lst:
        phase = np.average(comp.phase_array[comp.x1:comp.x2])
        
        ca_si_inds = np.where((comp.wavelength >= ca_si_range[0]) & (comp.wavelength <= ca_si_range[1]))
        si_inds = np.where((comp.wavelength >= si_range[0]) & (comp.wavelength <= si_range[1]))
        nir_trip_inds = np.where((comp.wavelength >= nir_trip_range[0]) & (comp.wavelength <= nir_trip_range[1]))

        ca_si = plt.subplot(gs[0])
        ca_si.axes.get_yaxis().set_ticks([])
        ca_si.axes.set_title("Ca/Si", fontdict = font2, fontsize = 15)
        plt.xticks(rotation=50, fontsize = 10)
        ca_si.axes.xaxis.set_tick_params(length=5, width = 2)
#         ca_si.axes.xaxis.set_ticks(np.arange(ca_si_range[0], ca_si_range[1], 1000))
        plt.plot(comp.wavelength[ca_si_inds], 5.e15*comp.flux[ca_si_inds]- 9*i, 'b-', linewidth = 4, color = s_m.to_rgba(phase))
        plt.ylabel('Relative Flux + const.', fontdict = font1, fontsize = 20)
        
        si = plt.subplot(gs[1])
        si.axes.get_yaxis().set_ticks([])
        si.axes.set_title("Si", fontdict = font2, fontsize = 15)
        plt.xticks(rotation=50, fontsize = 10)
        si.axes.xaxis.set_tick_params(length=5, width = 2)
#         si.axes.xaxis.set_ticks(np.arange(si_range[0], si_range[1], 1000))
        plt.plot(comp.wavelength[si_inds], 7.e15*comp.flux[si_inds] - 4.5*i, 'g-', linewidth = 4, color = s_m.to_rgba(phase))
    
        nir_trip = plt.subplot(gs[2])
        nir_trip.axes.get_yaxis().set_ticks([])
        nir_trip.axes.set_title("NIR Triplet", fontdict = font2, fontsize = 15)
        plt.xticks(rotation=50, fontsize = 10)
        nir_trip.axes.xaxis.set_tick_params(length=5, width = 2)
#         nir_trip.axes.xaxis.set_ticks(np.arange(nir_trip_range[0], nir_trip_range[1], 1000))
        plt.plot(comp.wavelength[nir_trip_inds], 6.e15*comp.flux[nir_trip_inds] - 3*i, 'r-', linewidth = 4, color = s_m.to_rgba(phase))
        
        i+=1
        
    cb = plt.colorbar(s_m, ax = fig.axes, aspect=50)
    cb.set_label('Phase', fontdict = font2, fontsize = 15)
    cb.ax.tick_params(labelsize=10)
    cb.ax.invert_yaxis()
    plt.savefig('../../Paper_Drafts/zoom_in.png', dpi = 300, bbox_inches = 'tight')
    plt.show()
```

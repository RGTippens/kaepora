```python
composites = query_db.main(len(query_strings), query_strings)
```

```python
def stacked_plot(composites):
    font1 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 20,
        }
    
    font2 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 10,
        }
    
    fig, ax = plt.subplots(1,1)
    ax.get_yaxis().set_ticks([])
    plt.ylabel('Relative Flux + const.', fontdict = font1)
    plt.xlabel('Wavelength (A)', fontdict = font1)
    fig.set_size_inches(15., 30.)
    composites, scales = composite.optimize_scales(composites, composites[0], True)

    for comp in composites:
        phase = np.average(comp.phase_array[comp.x1:comp.x2])
        ax.plot(comp.wavelength[comp.x1:comp.x2], 2.e15*comp.flux[comp.x1:comp.x2] - 3.5*phase, 'b-', linewidth = 1)
        plt.fill_between(comp.wavelength[comp.x1:comp.x2], 2.e15*comp.low_conf[comp.x1:comp.x2]- 3.5*phase, 2.e15*comp.up_conf[comp.x1:comp.x2] - 3.5*phase, alpha = 0.5)
        plt.text(10300, comp.flux[comp.x2] - 3.5*phase + 1, str(round(phase, 2)) + ' days', fontdict = font2)
#     plt.savefig('../../FOE_figures/all_t_spec.png', dpi = 100, bbox_inches = 'tight')
    plt.show()
```

## Zoom In Plots

```python
import matplotlib.gridspec as gridspec
```

```python
def zoom_in_plot(lst):
    font1 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 20,
        }
    
    font2 = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 15,
        }
    
    ca_si_range = [3400., 4400.]
    si_range = [5500., 6500.]
    nir_trip_range = [7500., 9000]
    
    gs = gridspec.GridSpec(1, 3)
    fig = plt.figure(num = 1, dpi = 100, figsize = [8,12], facecolor = 'w')
    fig.text(0.5, 0.04, 'Wavelength ($\AA$)', ha='center', fontdict = font1)
    s_m = query_db.make_colorbar(lst)
    
    for comp in lst:
        phase = np.average(comp.phase_array[comp.x1:comp.x2])
        
        ca_si_inds = np.where((comp.wavelength >= ca_si_range[0]) & (comp.wavelength <= ca_si_range[1]))
        si_inds = np.where((comp.wavelength >= si_range[0]) & (comp.wavelength <= si_range[1]))
        nir_trip_inds = np.where((comp.wavelength >= nir_trip_range[0]) & (comp.wavelength <= nir_trip_range[1]))

        ca_si = plt.subplot(gs[0])
        ca_si.axes.get_yaxis().set_ticks([])
        plt.xticks(rotation=50)
#         ca_si.axes.xaxis.set_ticks(np.arange(ca_si_range[0], ca_si_range[1], 1000))
        plt.plot(comp.wavelength[ca_si_inds], 5.e15*comp.flux[ca_si_inds]- 9*phase, 'b-', linewidth = 2, color = s_m.to_rgba(phase))
        plt.ylabel('Relative Flux + const.', fontdict = font1)
        
        si = plt.subplot(gs[1])
        si.axes.get_yaxis().set_ticks([])
        plt.xticks(rotation=50)
#         si.axes.xaxis.set_ticks(np.arange(si_range[0], si_range[1], 1000))
        plt.plot(comp.wavelength[si_inds], 7.e15*comp.flux[si_inds] - 4.5*phase, 'g-', linewidth = 2, color = s_m.to_rgba(phase))
    
        nir_trip = plt.subplot(gs[2])
        nir_trip.axes.get_yaxis().set_ticks([])
        plt.xticks(rotation=50)
#         nir_trip.axes.xaxis.set_ticks(np.arange(nir_trip_range[0], nir_trip_range[1], 1000))
        plt.plot(comp.wavelength[nir_trip_inds], 6.e15*comp.flux[nir_trip_inds] - 3*phase, 'r-', linewidth = 2, color = s_m.to_rgba(phase))
    
    cb = plt.colorbar(s_m, ax = fig.axes, aspect=50)
    cb.set_label('Phase', fontdict = font2)
    cb.ax.invert_yaxis()
#     plt.savefig('../../FOE_figures/zoom_in.png', dpi = 100, bbox_inches = 'tight')
    plt.show()
```

```python
zoom_list = composites[0:26]
zoom_in_plot(zoom_list)
```
